### Задание 2

```python
t = np.linspace(0, 10, num=51)
f = np.cos(t)

plt.plot(t, f, color='lightgreen')

plt.xlabel("Значения t")
plt.ylabel("Значения f")
plt.title('График f(t)')
plt.xlim([0.5, 9.5])
plt.ylim([-2.5, 2.5])

plt.show()
```

### Задание 3

```python
x = np.linspace(-3, 3, num=51)
y1 = x ** 2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)

fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

fig.set_size_inches(10, 8) # я сделала числа побольше потому что при 8, 6 графики все равно были маленькими
fig.subplots_adjust(wspace=0.3, hspace=0.3)

# заметка для себя))) вместо ax1 можно:
# ax[0, 0].plot(x, y1)
# ax[0, 0].set_title("x^2") итд

ax1.plot(x, y1, color='mediumorchid')
# не поняла про ограничение по оси x [-5, 5], если у нас [-3, 3]
ax1.set_title("x^2")

ax2.plot(x, y2, color='coral')
ax2.set_title("2x + 0.5")

ax3.plot(x, y3, color='seagreen')
ax3.set_title("-3x - 1.5")

ax4.plot(x, y4, color='royalblue')
ax4.set_title("sin(x)")

plt.show()
```
